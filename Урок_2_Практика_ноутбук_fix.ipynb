# Практика 2: Условные конструкции и циклы

Условные конструкции и циклы в Python являются важными инструментами для анализа данных. Они позволяют создавать программы, которые могут принимать решения на основе определенных условий и повторять указанные действия определенное количество раз.

Условные конструкции используются для определения действий, которые должны быть выполнены в зависимости от результата проверки условия. В Python есть несколько типов условных конструкций, включая `if, elif и else`.

Циклы же используются для повторения блока кода определенное количество раз. В Python есть несколько типов циклов, включая `while и for`.

Освоение этих инструментов позволяет аналитикам данных эффективно обрабатывать большие объемы информации, автоматизировать рутинные задачи и создавать сложные алгоритмы анализа данных.

```python
# Задаем переменную
num = 10

if num > 0:
    print("Число положительное")
    # Во вложенной условной конструкции дополнительно смотрим число четное или нет
    if num % 2 == 0:
        print("Число четное")
    else:
        print("Число нечетное")
else:
    print("Число отрицательное или равно нулю")
      
# Вывод
Число положительное
Число четное
```       

Посчитайте, сколько всего уникальных курсов в выгруженном списке `courses`, затем в зависимости от их числа выведите сообщение:

* если в списке больше 7 уникальных наименований, выведите `Прав Костя.`
* если в списке их ровно 7, выведите `Права Настя.`
* если их меньше 7, то сообщите ребятам, что они оба неправы и напечайте `Оба менеджера неправы, у нас N специализаций.`, где на месте `N` укажите число уникальных специализаций

````python
courses = [
    'Start DA', 'Start ML', 'Start ML', 'Hard ML', 'Start DA', 'Hard ML',
    'Hard DA', 'Start DE', 'Start DA', 'Start DE', 'Start DL', 'Hard DA',
    'Start DE', 'Start DA', 'Start DE', 'Hard ML', 'Start ML', 'Hard ML'
    ]
print(set(courses))

if len(set(courses)) > 7:
    print('Прав Костя.')
elif len(set(courses)) == 7:
    print('Права Настя.')
elif len(set(courses)) < 7:
    print(f'Оба менеджера неправы, у нас {len(set(courses))} специализаций.')  
````

Ключи в словаре — это, судя по всему, логины студентов, а значения по ключам — ещё одни словари с данными студентов и их оценками. В свою очередь, эти вложенные словари содержат ключи `'name'`, `'surname'`, `'points'`, по которым находятся, соответственно, имена, фамилии и списки с баллами студентов.

Из инструкции вы также поняли, что в уроке всего 5 заданий и финальный проект. По ключу `'points'` как раз указан список с баллами, в котором всегда первые 5 значений — это оценки за 5 заданий в уроке, при этом за каждое начисляется от 0 до 5 баллов. Последнее значение в списке — это всегда оценка финального проекта в уроке, за который можно получить от 0 до 10 баллов.

Считается, что студент успешно прошёл урок, если его средний балл за 5 заданий больше 3, а оценка за финальный проект не меньше 5 баллов.

Для каждого студента в словаре `student_points` необходимо определить, удалось ли ему успешно пройти урок.

Результат нужно представить в виде словаря `student_results`, где по ключу `<Имя Фамилия>` будут словари с ключами `'points_avg'`, `'points_final'` и `'result'`. По этим ключам необходимо указать, соответственно, средний балл за 5 заданий, оценку за финальный проект и результат `'passed'` или `'failed'` в зависимости от оценок.

**Пример**:

Если студент `Иван Иванов` имеет оценки `[5, 5, 5, 5, 5, 10]`, то в словарь `student_results` должна попасть следующая пара ключ-значение:

`'Иван Иванов': {'points_avg': 5.0, 'points_final': 10, 'result': 'passed'}`

Список `emails` уже создан:

````python
student_data = {
    'alice.dem': {'name': 'Алиса', 'surname': 'Демидова', 'points': [5, 5, 5, 5, 5, 10]},
    'mikhail.val': {'name': 'Михаил', 'surname': 'Валетов', 'points': [3, 3, 5, 1, 4, 8]},
    'anton.kar': {'name': 'Антон', 'surname': 'Каргин', 'points': [5, 1, 5, 2, 0, 5]},
    'maria.vol': {'name': 'Мария', 'surname': 'Волкова', 'points': [4, 4, 5, 4, 3, 7]},
    'roman.tok': {'name': 'Роман', 'surname': 'Токарев', 'points': [2, 3, 5, 3, 3, 0]},
}

student_results = {}
for key, value in student_data.items():
    name = value['name']
    surname = value['surname']
    full_name = f"{name} {surname}"
    points = value['points']
    cres_point = points[:5]
    point_avg = sum(cres_point) / len(cres_point)
    points_final = points[-1]
    
    if point_avg > 3 and points_final >= 5:
        result = 'passed'
    else:
        result = 'failed'
    student_results[full_name] = {
        'points_avg': point_avg,
        'points_final': points_final,
        'result': result
    }
print(student_results)
````

````python
# Создаем словарь для сбора итоговой информации
student_results = {}
# Запускаем цикл по словарю student_data
for student in student_data:
    # В переменную points_all сохраняем список всех оценок студентов
    points_all = student_data[student]['points']
    # В переменную points_tasks сохраняем список оценок за задания
    points_tasks = points_all[:len(points_all)-1]
    # В переменную points_final сохраняем оценки за итоговый тест
    points_final = points_all[-1]
    # Вычисляем средную оценку студента и сохраняем в переменную points_tasks_avg
    points_tasks_avg = sum(points_tasks) / len(points_tasks)
    # Проверяем условия для прохождения урока
    if points_tasks_avg > 3 and points_final >= 5:
        result = 'passed'
    else:
        result = 'failed'
    # Формируем в одну строку имя и фамилию студента и сохраняем в переменную student_name
    student_name = student_data[student]['name'] + ' ' + student_data[student]['surname']
    # Передаем все данные в итоговый словарь 
    student_results[student_name] = {'points_avg': points_tasks_avg, 'points_final': points_final, 'result': result}
````
